```python
import pandas as pd
import ast
import jsonlines
import random

# Read the CSV file into a DataFrame
read_results_df = pd.read_csv("image_classes_infilled.csv")

newvar = read_results_df["Class_Frequencies"].apply(ast.literal_eval)
image_files = read_results_df["Image"]

jsonl_file_path = "./hateful_memes/train.jsonl"

labels_1 = []
labels_0 = []

# Read the jsonl file and extract labels
with jsonlines.open(jsonl_file_path) as reader:
    for i, entry in enumerate(reader):
        label = entry['label']
        if label == 1:
            labels_1.append((entry, i))
        elif label == 0:
            labels_0.append((entry, i))

# Randomly select 2500 labels with value 1
selected_labels_1 = random.sample(labels_1, k=2500)

# Randomly select 2500 labels with value 0
selected_labels_0 = random.sample(labels_0, k=2500)

# Combine the selected labels
selected_labels = selected_labels_1 + selected_labels_0

# Shuffle the combined list to ensure randomness
random.shuffle(selected_labels)

# Display the first few selected labels
print(selected_labels[0])

# print(newvar)
```

```python
results = {}

for index, label in selected_labels:
    # Extract information from newvar using the index
    # newvar[index]

    frequencies_dict = newvar[label]
    
    # # Perform your logic using the extracted information
    for key, value in frequencies_dict.items():
        if index['label'] == 1:
            results[key] = results.get(key, 0) + value
        elif index['label'] == 0:
            results[key] = results.get(key, 0) - value

for i in results.keys():
    if results[i] > 0:
        results[i] = 1
    elif results[i] < 0:
        results[i] = -1
    else:
        results[i] = 0

results_df = pd.DataFrame(list(results.items()), columns=["Class", "Result"])

results_df.to_csv("train_results.csv", index=False)

results
```
